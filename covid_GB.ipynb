```python
# Check for non-numeric columns
non_numeric_columns = df_filled.select_dtypes(exclude=['number']).columns
print("Non-numeric columns: ", non_numeric_columns)

df_filled[non_numeric_columns] = df_filled[non_numeric_columns].apply(pd.to_numeric, errors='coerce')


df_filled['Date'] = pd.to_datetime(df_filled['Date'], errors='coerce')

# Drop rows where 'Date' could not be converted to datetime (if any)
df_filled = df_filled.dropna(subset=['Date'])


# Now resample weekly and monthly
weekly_df = df_filled.resample('W', on='Date').sum()
monthly_df = df_filled.resample('M', on='Date').sum()

# Save weekly and monthly data
weekly_df.to_excel('Weekly_Stats_GB.xlsx')
monthly_df.to_excel('Monthly_Stats_GB.xlsx')

weekly_df.head(), monthly_df.head()
```

```python
# Weekly Max/Min
weekly_max_tests = weekly_df['Daily Tests'].max()
weekly_min_tests = weekly_df['Daily Tests'].min()

# Monthly Max/Min
monthly_max_tests = monthly_df['Daily Tests'].max()
monthly_min_tests = monthly_df['Daily Tests'].min()

weekly_max_tests, weekly_min_tests, monthly_max_tests, monthly_min_tests
```

```python
#q4

# Assuming 'data_filled' is your DataFrame
population_million = 220

# Convert the 'Date' column to datetime format if not already
df_filled['Date'] = pd.to_datetime(df_filled['Date'])

# Calculate Per Thousand and Per Million stats
df_filled['Tests per Thousand'] = (df_filled['Daily Tests'] / population_million) / 1000
df_filled['Cases per Thousand'] = (df_filled['Daily Cases'] / population_million) / 1000
df_filled['Recoveries per Thousand'] = (df_filled['Daily Recoveries'] / population_million) / 1000
df_filled['Deaths per Thousand'] = (df_filled['Daily Deaths'] / population_million) / 1000

df_filled['Tests per Million'] = (df_filled['Daily Tests'] / population_million) * 1
df_filled['Cases per Million'] = (df_filled['Daily Cases'] / population_million) * 1
df_filled['Recoveries per Million'] = (df_filled['Daily Recoveries'] / population_million) * 1
df_filled['Deaths per Million'] = (df_filled['Daily Deaths'] / population_million) * 1

# Group by Month and aggregate the data
monthly_data = df_filled.groupby(df_filled['Date'].dt.to_period('M')).agg({
    'Daily Tests': 'sum',
    'Daily Cases': 'sum',
    'Tests per Thousand': 'sum',
    'Cases per Thousand': 'sum',
    'Tests per Million': 'sum',
    'Cases per Million': 'sum',
}).reset_index()

# Rename the 'Date' column to 'Month' and format it
monthly_data['Month'] = monthly_data['Date'].dt.strftime('%B')
monthly_data = monthly_data.drop(columns=['Date'])

# Calculate the positive rate
monthly_data['Positive Rate'] = (monthly_data['Daily Cases'] / monthly_data['Daily Tests']) * 100

# Rearrange the columns to match the format in the image
monthly_data = monthly_data[['Month', 'Daily Tests', 'Daily Cases', 'Tests per Thousand', 'Cases per Thousand', 'Tests per Million', 'Cases per Million', 'Positive Rate']]

# Display the DataFrame as a formatted table
styled_table = monthly_data.style.set_table_styles(
    [{'selector': 'thead th', 'props': [('background-color', '#2d4154'), ('color', 'white')]}]
).background_gradient(cmap='Blues', subset=['Daily Tests', 'Daily Cases', 'Tests per Thousand', 'Cases per Thousand', 'Tests per Million', 'Cases per Million', 'Positive Rate'])

# Show the styled DataFrame
styled_table
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Prepare data for ML models
X = df_filled[['Daily Tests']]  # Feature (you can choose other features as well)
cases = df_filled['Daily Cases']  # Target for cases
deaths = df_filled['Daily Deaths']  # Target for deaths

# Drop rows with missing values in X, y_cases, or y_deaths
df_clean = df_filled.dropna(subset=['Daily Tests', 'Daily Cases', 'Daily Deaths'])
X_clean = df_clean[['Daily Tests']]
cases_clean = df_clean['Daily Cases']
deaths_clean = df_clean['Daily Deaths']

# Split the data into training and testing sets
X_train, X_test, y_cases_train, y_cases_test = train_test_split(X_clean, cases_clean, test_size=0.2, random_state=42)
_, _, y_deaths_train, y_deaths_test = train_test_split(X_clean, deaths_clean, test_size=0.2, random_state=42)

# Linear Regression Model for Daily Cases
model = LinearRegression()
model.fit(X_train, y_cases_train)
cases_predictions = model.predict(X_test)


# Display the results
print(f"Cases: {cases_predictions}")
```

```python

# Random Forest Model for Daily Deaths
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_deaths_train)
deaths_predictions = rf_model.predict(X_test)

print(f" Deaths: {deaths_predictions}")
```
